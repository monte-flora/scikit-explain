This notebook demostrates using Sobol Indices in `scikit-explain` for diagnosing feature interactions. 

```python
import sys, os 
sys.path.insert(0, os.path.dirname(os.getcwd()))

import skexplain
import plotting_config 
```

```python
### Loading the training data and pre-fit models 
estimators = skexplain.load_models()
X,y = skexplain.load_data()
```

```python
explainer = skexplain.ExplainToolkit(estimators[0], X=X, y=y)
results = explainer.sobol_indices(n_bootstrap=5000, class_index=1)
```

```python
explainer.plot_importance(data=[results, results, results], 
                          panels=[('sobol_total', 'Random Forest'), ('sobol_1st', 'Random Forest'), 
                                 ('sobol_interact', 'Random Forest')],
                          display_feature_names = plotting_config.display_feature_names,
                          figsize=(12,4)
                         )
```
